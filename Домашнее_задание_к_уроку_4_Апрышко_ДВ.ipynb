```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('kc_house_data.csv', sep=',')
df
```

### 1.2 Изучите стоимости недвижимости

```python
plt.figure(figsize=(12, 6))
plt.hist(df['price'], bins=50)
plt.title('Распределение стоимости недвижимости')
plt.xlabel('Стоимость')
plt.ylabel('Количество');
```

### 1.3 Изучите распределение квадратуры жилой

```python
plt.figure(figsize=(12, 8))
sns.histplot(df['sqft_living'], bins=50)
plt.title('Распределение квадратуры жилой')
plt.xlabel('Кв. метры жилые')
plt.ylabel('Количество');
```

Вывод: Подавляющее большинство домов имеет жилую пложадь от 1000 до 2500 кв. метров. Количество домов с жилой площадью свыше 5000 кв. метров незначительное. Единичные экземпляры имеют жилую площадь более 6000 кв. метров. График распределения количества домов по жилой площади повторяет график распределения количества домов по цене. Это говорит о прямой зависимости цены домов от их жилой площади.

### 1.4 Изучите распределение года постройки

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['yr_built'], bins=58)
plt.title('Распределение года постройки')
plt.xlabel('Год постройки')
plt.ylabel('Количество');
```

Вывод: Домов старше 1940 года сильно меньше, чем домов более молодых. Это может быть следствием того, что более старые дома вытесняются более новыми, или начиная с этого времени начали строить больше домов.

### 2.1 Изучите распределение домов от наличия вида на набережную

```python
data = df['waterfront'].value_counts()

plt.figure(figsize=(4, 4))
plt.pie(data, autopct='%1.1f%%')
plt.legend(['no waterfront', 'has waterfront']);
```

### 2.2 Изучите распределение этажей домов

```python
data = df['floors'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(data, autopct='%1.1f%%', labels=data.index)
plt.title('Количество этажей');
```

Вывод: Около половины всех домов - одноэтажные. Менее 5% домов выше 2 этажей.

### 2.3 Изучите распределение состояния домов

```python
data = df['condition'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(data, autopct='%1.1f%%', labels=data.index);
```

Вывод: Около 65% домов имеют удовлетворительное состояние. Хорошее и отличное состояние имеют около трети домов. Неудовлетворительное состояние у менее чем 1% домов.

## Условие 3: 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df.corr(numeric_only=True)
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='PuRd');
```

Вывод: Цена сильно зависит от жилой площадиоценки дома, и в меньшей степени от количества спален и ванных. Интересно, что цена не зависит от года постройки или обновления.

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg');
```

Вывод: Стоимость площадь дома напрямую влияет на цену. Чем больше жилая площадь дома, тем выше его стоимость. При этом с ростом площади растет как минимальная цена, так и диапазон цен на сходные по жилой площади дома.

```python
sns.jointplot(x=df['price'], y=df['grade'], kind='reg');
```

Вывод: Оценка дома напрямую влияет на цену. Чем выше оценка, тем дороже дома. При этом видно, что чем выше оценка домов, тем выше как минимальная цена, так и шире диапазон цен домов.

```python
sns.boxplot(x=df['price'], y=df['grade'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('grade')
plt.title('Distribution of price by grade');
```

Вывод: Оценка дома напрямую и достаточно сильно влияют на стоимость дома. Дома с оценкой 11-12-13 имеют наибольшую стоимость.

```python
sns.boxplot(x=df['price'], y=df['floors'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('floors')
plt.title('Distribution of price by floors');
```

Вывод: Этажность домов не напрямую влияет на цену дома. Выделяются дома с количеством этажей 2.5, которые имеют наибольшую стоимость. 

```python
sns.boxplot(x=df['price'], y=df['view'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('view')
plt.title('Distribution of price by view');
```

Вывод: Оценка вида существенно влияет на стоимость домов. Наивысшую цену в среднем имеют дома с оценков вида 4.
